```python
def density_plot(feature_var, class_var):
    sns.set_style(style="whitegrid")
    sns.FacetGrid(data=haber, hue=class_var, size=4) \
    .map(sns.distplot, feature_var) \
    .add_legend()
```

```python
density_plot('age','status')
plt.show()
```
